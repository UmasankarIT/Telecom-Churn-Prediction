MULTI VARIATE ANALYSIS - studies the relationship between three or more variables at the same time.

```python
# Year + Month + Value (Seasonal Trends over Years) - Shows how subscribers change by month across years.
# without using .unstack()

monthly_yearly = df.groupby(['year', 'month'])['value'].sum()
print(monthly_yearly.head())
```

```python
df.head()
```

UNSTACK() :

When you do a groupby() with two or more columns, you get a multi-level index (hierarchical) DataFrame.

The .unstack() function converts one of those levels into columns  making your data table wider instead of taller.

.stack()  makes your data longer (more rows), moves columns -> rows

.unstack()  makes your data wider (more columns), moves rows -> columns

-------------------------------------------------------------------------------

Year + Month + Value (Seasonal Trends over Years) --> Shows how subscribers change by month across years.

```python
# Year + Month + Value (Seasonal Trends over Years) - Shows how subscribers change by month across years.

monthly_yearly = df.groupby(['year', 'month'])['value'].sum().unstack()

monthly_order = ['January','February','March','April','May','June','July','August','September','October','November','December']
monthly_yearly = monthly_yearly[monthly_order]

print(monthly_yearly.head())
```

```python
monthly_yearly.plot(figsize=(12,6))
plt.title("Monthly Subscriber Trends Over Years")
plt.xlabel("Year")
plt.ylabel("Subscribers")
plt.legend(title="Month")
plt.show()
```

Circle + Year + Value (Geography Trends Over Time) --> Shows how subscriber counts change for each circle over years.

```python
circle_yearly.head(10)
```

Service Provider + Year + Value --> Shows growth of each provider over years.

```python
provider_yearly = df.groupby(['service_provider','year'])['value'].sum().unstack()
provider_yearly = provider_yearly.fillna(0)
print(provider_yearly.head())
```

-------------------------------------------------------------------------------

Circle + Service Provider + Value --> Shows which provider has most subscribers in which circle.

```python
circle_provider = df.groupby(['circle','service_provider'])['value'].sum().unstack()
circle_provider = circle_provider.fillna(0)
print(circle_provider.head(1))
```

```python
circle_provider.plot(kind='bar', stacked=True, figsize=(15,7))
plt.title("Subscribers by Circle and Service Provider")
plt.xlabel("Circle")
plt.ylabel("Subscribers")
plt.legend(title="Service Provider", bbox_to_anchor=(1.05, 1))
plt.show()
```

Connection Type + Circle + Value -- > Shows which type of connection is more dominant in each circle.

```python
type_circle = df.groupby(['circle','type_of_connection'])['value'].sum().unstack()
type_circle = type_circle.fillna(0)
print(type_circle.head(3))
```

```python
type_circle.plot(kind='bar', stacked=True, figsize=(15,7))
plt.title("Subscribers by Circle and Connection Type")
plt.xlabel("Circle")
plt.ylabel("Subscribers")
plt.legend(title="Connection Type", bbox_to_anchor=(1.05, 1))
plt.show()
```
